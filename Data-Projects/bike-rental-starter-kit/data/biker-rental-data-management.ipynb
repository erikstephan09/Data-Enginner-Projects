### Creating a data table

* Observation
    - Our data table will have year, month, day

```python
#we use datatime library
datetime_init = datetime.datetime(2016,1,1)
data_colum = [datetime_init + datetime.timedelta(x) for x in range(366)]
#print(data_colum)
df_date = pd.DataFrame(data_colum, dtype = 'datetime64[ns]')
df_date.columns = ['date']
df_date.head(10)
```

We will include the next columns: date_key year, month and day. date_key will be our primary key 

```python
df_date['id'] = df_date.index
df_date['date_key'] = df_date['date'].apply(lambda x: int(x.strftime('%Y%m%d')))
df_date['year'] = df_date['date'].apply(lambda x: x.year)
df_date['month'] = df_date['date'].apply(lambda x: x.month)
df_date['day'] = df_date['date'].apply(lambda x: x.day)
df_date.head()
```

Now we will create a data_key column in bike table and weather table

```python
df['date_key'] = df['start_time'].apply(lambda x: int(x.strftime('%Y%m%d')))
df_w['date_key'] = df_w['date'].apply(lambda x: int(x.strftime('%Y%m%d')))
```

### start_station table

```python
start_station = pd.DataFrame({
    'id' : None,
    'start_id' : df['start_station_id'].copy(True),
    'start_station_name' : df['start_station_name'].copy(True),
    'start_station_latitude' : df['start_station_latitude'].copy(True),
    'start_station_latitude' : df['start_station_latitude'].copy(True),
    'start_station_longitude' : df['start_station_longitude'].copy(True)
})
start_station.reset_index(drop = True, inplace = True)
start_station['id'] = start_station.index
```

```python
start_station.head()
```

We have duplicated information, we will drop some rows

```python
start_station.drop_duplicates(subset = ['start_id'], inplace = True)
start_station.reset_index(drop = True, inplace = True)
start_station['id'] = start_station.index
```

```python
start_station.head()
```

### end_station table

```python
end_station = pd.DataFrame({
    'id' : None,
    'end_id' : df['end_station_id'].copy(True),
    'end_station_name' : df['end_station_name'].copy(True),
    'end_station_latitude' : df['end_station_latitude'].copy(True),
    'end_station_longitude' : df['end_station_longitude'].copy(True),
})
end_station.reset_index(drop = True, inplace = True)
end_station['id'] = end_station.index
```

We need to drop de duplicated values to

```python
end_station.drop_duplicates(subset = ['end_id', 'end_station_name', 'end_station_latitude', 'end_station_longitude'], inplace = True)
end_station.reset_index(drop = True, inplace = True)
end_station['id'] = end_station.index
```
